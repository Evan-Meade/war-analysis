```python
wg.reset()
wg.play_game()
wg.turn_data
```

```python
war_game_data = pd.DataFrame()
war_turn_data = pd.DataFrame()
n = 10
for i in range(n):
    wg.reset()
    wg.play_game()
    game_data = wg.game_dict()
    game_df = pd.DataFrame(game_data, index=[i])
#     game_df.at[i, 'player0_start.hand'] = wg.start0
#     game_df.at[i, 'player1_start.hand'] = wg.start1
    war_game_data = pd.concat([war_game_data, game_df])
    
    war_turn_data = pd.concat([war_turn_data, wg.turn_data])
war_game_data.to_csv('war_game_data.csv', index=False)
war_turn_data.to_csv('war_turn_data.csv', index=False)

war_turn_data.iloc[40:70, :]
```
